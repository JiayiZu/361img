```python
import pandas as pd
import numpy as np
import math
from fontTools.subset import subset
```

```python
def calculate_entropy(dataset):
    rlt = 0
    a_dict = {}
    for row in dataset:
        if not row[0] in a_dict.keys():
            a_dict[row[0]] = 0
        a_dict[row[0]] += 1
    for num in a_dict.values():
        p = num / len(dataset)
        rlt -= p * math.log(p, 2)
    return rlt
```

```python
def choose_feature(dataset):
    ig = 0
    index = -1
    entropy = calculate_entropy(dataset)
    for i in range(1, len(dataset[0])):
        e = 0
        feature = [row[i] for row in dataset]
        features_set = set(feature)
        for value in features_set:
            sub_dataset = set_data(dataset, i, value)
            weight = len(sub_dataset) / float(len(dataset))
            e += weight * calculate_entropy(sub_dataset)
        temp = entropy - e
        if temp > ig:
            ig = temp
            index = i
    return index
```

```python
def set_data(dataset, axis, value):
    rlt = []
    for row in dataset:
        if row[axis] == value:
            reduced_featvec = row[:axis]
            reduced_featvec.extend(row[axis + 1:])
            rlt.append(reduced_featvec)
    return rlt
```

```python
def majority_label(types):
    a_dict = {}
    for type in types:
        if type not in a_dict.keys():
            a_dict[type] = 0
        a_dict[type] += 1
    sort_dict = sorted(a_dict.items(), key=lambda x: x[1], reverse=True)
    return sort_dict[0][0]
```

```python
def build_tree(dataset, labels, level=-99):
    if level == 0:
        return None
    types_arr = [row[0] for row in dataset]
    if types_arr.count(types_arr[0]) == len(types_arr):
        return types_arr[0]
    if len(dataset[0]) == 1:
        return majority_label(types_arr)
    index = choose_feature(dataset)
    feature = labels[index]
    labels.append(feature)
    tree = {feature: {}}
    del (labels[index])
    values = set()
    {values.add(row[index]) for row in dataset}
    for value in values:
        temp_label = labels
        tree[feature][value] = build_tree(set_data(dataset, index, value), temp_label, level-1)
    return tree
```

```python
def predict(decision_tree, data, labels, f=None):
    for feature in decision_tree.keys():
        if feature in labels:
            index = labels.index(feature)
            rlt = data.iloc[0, index]
            if isinstance(decision_tree[feature], dict):
                if isinstance(decision_tree[feature][rlt], dict):
                    return predict(decision_tree[feature][rlt], data, labels, rlt)
                return decision_tree[feature][rlt]
            return decision_tree[feature]
        else:
            if isinstance(decision_tree[f], dict):
                return predict(decision_tree[f], data, labels)
            return decision_tree[f]
    return None
```

```python
def test_decision_tree(decision_tree, data, labels):
    correct = 0
    for row in data:
        if predict(decision_tree, pd.DataFrame([row], columns=range(len(row))), labels) == row[0]:
            correct += 1
    accuracy = correct / len(data)
    return accuracy
```
